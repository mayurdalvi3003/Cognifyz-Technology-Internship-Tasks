# Task1: Restaurant Ratings

### Analyze the distribution of aggregate ratings and determine the most common rating range.

```python
df['Aggregate rating'].value_counts().plot(kind ='bar', title = 'Distribution of Aggregate ratings' , xlabel = 'Aggregate rating', ylabel = 'count')
```

```python
most_common_rating = df['Aggregate rating'].value_counts().idxmax()
print(f'Most Common Rating is : {most_common_rating}')
```

###  Calculate the average number of votes received by restaurants.


```python
avg_votes = round(df['Votes'].mean())
print(f'The average number of votes received by restaurants is: {avg_votes}')
```

#  Task2: Cuisine Combination

###  Identify the most common combinations of cuisines in the dataset

```python
most_common_combinations = df['Cuisines'].value_counts().head(10)
most_common_combinations
```

### Determine if certain cuisine combinations tend to have higher ratings

```python
name = df.groupby("Cuisines")["Aggregate rating"].mean().sort_values(ascending = False).head(1).index[0]
avg = df.groupby("Cuisines")["Aggregate rating"].mean().sort_values(ascending = False).head(1).values[0]
```

```python
print(f'{name} is having {avg} average rating')
```

```python
import folium
import networkx as nx

# Example data: nodes with geographical coordinates and parameters
nodes = {
    'A': {'pos': (34.0522, -118.2437), 'parameter': 10},
    'B': {'pos': (40.7128, -74.0060), 'parameter': 20},
    'C': {'pos': (41.8781, -87.6298), 'parameter': 15}
}

# Example edges connecting nodes
edges = [('A', 'B'), ('A', 'C'), ('B', 'C')]

# Create a graph
G = nx.Graph()

# Add nodes with their attributes
for node, attr in nodes.items():
    G.add_node(node, pos=attr['pos'], parameter=attr['parameter'])

# Add edges
G.add_edges_from(edges)

# Create a Folium map
map_center = [sum(x[0] for x in nodes.values())/len(nodes), sum(x[1] for x in nodes.values())/len(nodes)]
mymap = folium.Map(location=map_center, zoom_start=4)

# Plot nodes
for node, data in G.nodes(data=True):
    folium.Marker(data['pos'], popup=f"Node {node}, Parameter: {data['parameter']}").add_to(mymap)

# Plot edges
for edge in G.edges():
    pos1 = nodes[edge[0]]['pos']
    pos2 = nodes[edge[1]]['pos']
    folium.PolyLine([pos1, pos2], color="blue", weight=2.5, opacity=1).add_to(mymap)

# Save the map
mymap.save("graph_visualization.html")
```
